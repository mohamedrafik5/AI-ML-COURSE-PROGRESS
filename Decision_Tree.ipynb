```python
import pandas as pd
```

```python
dataset = pd.read_csv("/content/data.csv")
```

```python
dataset.describe()
```

```python
dataset.info()
```

```python
dataset.isnull().sum()
```

```python
dataset.head()
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
le = LabelEncoder()
dataset['street_Encoded'] = le.fit_transform(dataset['street'])
dataset['city_Encoded'] = le.fit_transform(dataset['city'])
dataset['country_Encoded'] = le.fit_transform(dataset['country'])

dataset
```

```python
x = dataset[['street_Encoded','city_Encoded','country_Encoded','bedrooms','bathrooms','sqft_living','floors','waterfront','condition','yr_renovated']]
y = dataset['price']
```

```python
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
```

```python
from sklearn.tree import DecisionTreeRegressor
model = DecisionTreeRegressor()
model.fit(x_train, y_train)
```

```python
y_pred = model.predict(x_test)
y_pred
```

```python
from sklearn.metrics import r2_score, mean_squared_error
r2 = r2_score(y_test, y_pred)
print(f"R-squared: {r2}")
mse = mean_squared_error(y_test, y_pred)
print(f"Mean Squared Error: {mse}")
```

```python
import pickle
file_name = "House price model.sav"
pickle.dump(model, open(file_name, 'wb'))
```

```python
loaded_model = pickle.load(open(file_name, 'rb'))
```

```python
result =loaded_model.score(x_test, y_test)
print(result)
```

```python
example_input = [[1522, 36, 0, 3.0, 1.50, 1340, 1.5, 0, 3, 2005]]

result = loaded_model.predict(example_input)
print(result)
```

```python
import numpy as np
from sklearn.preprocessing import LabelEncoder
import pickle

file_name = "House price model.sav"
try:
    loaded_model = pickle.load(open(file_name, 'rb'))
except FileNotFoundError:
    print(f"Error: Model file '{file_name}' not found. Please train and save the model first.")
    loaded_model = None

if loaded_model:
    try:
        raw_street = input("Enter the street: ")
        raw_city = input("Enter the city: ")
        raw_country = input("Enter the country: ")
        bedrooms = float(input("Enter the number of bedrooms: "))
        bathrooms = float(input("Enter the number of bathrooms: "))
        sqft_living = int(input("Enter the square footage of living space: "))
        floors = float(input("Enter the number of floors: "))
        waterfront = int(input("Enter if waterfront (0 for No, 1 for Yes): "))
        condition = int(input("Enter the condition rating (1-5): "))
        yr_renovated = int(input("Enter the year renovated (0 if not renovated): "))
        dummy_data = {'street': ['18810 Densmore Ave N', '709 W Blaine St', '26206-26214 143rd Ave SE', raw_street],
                      'city': ['Shoreline', 'Seattle', 'Kent', raw_city],
                      'country': ['USA', 'USA', 'USA', raw_country]}
        dummy_df = pd.DataFrame(dummy_data)

        le_street = LabelEncoder()
        le_city = LabelEncoder()
        le_country = LabelEncoder()

        le_street.fit(dummy_df['street'])
        le_city.fit(dummy_df['city'])
        le_country.fit(dummy_df['country'])

        street_encoded = le_street.transform([raw_street])[0]
        city_encoded = le_city.transform([raw_city])[0]
        country_encoded = le_country.transform([raw_country])[0]

        example_input = [[street_encoded, city_encoded, country_encoded, bedrooms, bathrooms, sqft_living, floors, waterfront, condition, yr_renovated]]

        result = loaded_model.predict(example_input)
        print(f"\nPredicted house price: {result[0]:,.2f}")

    except ValueError:
        print("Invalid input. Please enter numeric values for bedrooms, bathrooms, sqft_living, floors, waterfront, condition, and yr_renovated.")
    except Exception as e:
        print(f"An error occurred: {e}")
```
